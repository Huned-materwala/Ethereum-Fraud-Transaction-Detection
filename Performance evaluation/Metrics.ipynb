```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to calculate mean and standard deviation
def calculate_stats(df, metrics_columns):
    means = df[metrics_columns].mean()
    stds = df[metrics_columns].std()
    return means, stds
```

```python
# List of CSV files and their corresponding metrics columns
Adaboost = r"../Learning models/Adaboost/ETFD_Adaboost.csv"
Decision_tree = r"../Learning models/Decision Tree/ETFD_DT.csv"
K_nearest_neighbor = r"../Learning models/K-Nearest Neighbor/ETFD_KNN.csv"
Random_forest = r"../Learning models/Random Forest/ETFD_RF.csv"
Stacking_mnb_rf = r"../Learning models/Stacking_MNB_RF/ETFD_Stacking_MNB_RF.csv"
Stacking_rf_lr = r"../Learning models/Stacking_RF_LR/ETFD_Stacking_RF_LR.csv"
Support_vector_machine = r"../Learning models/Support Vector Machine/ETFD_SVM.csv"
XGboost = r"../Learning models/XGboost/ETFD_XGBoost.csv"

Files = [Adaboost, Decision_tree, K_nearest_neighbor, Random_forest, Stacking_mnb_rf, Stacking_rf_lr, Support_vector_machine, XGboost]

metrics_columns = {
    Adaboost: ['Train Accuracy', 'Train Macro Precision', 'Train Macro Recall', 'Train Macro F1-Score', 'Train AUC', 'Test Accuracy', 'Test Macro Precision', 'Test Macro Recall', 'Test Macro F1-Score', 'Test AUC'],
    Decision_tree: ['Train Accuracy', 'Train Macro Precision', 'Train Macro Recall', 'Train Macro F1-Score', 'Train AUC', 'Test Accuracy', 'Test Macro Precision', 'Test Macro Recall', 'Test Macro F1-Score', 'Test AUC'],
    K_nearest_neighbor: ['Train Accuracy', 'Train Macro Precision', 'Train Macro Recall', 'Train Macro F1-Score', 'Train AUC', 'Test Accuracy', 'Test Macro Precision', 'Test Macro Recall', 'Test Macro F1-Score', 'Test AUC'],
    Random_forest: ['Train Accuracy', 'Train Macro Precision', 'Train Macro Recall', 'Train Macro F1-Score', 'Train AUC', 'Test Accuracy', 'Test Macro Precision', 'Test Macro Recall', 'Test Macro F1-Score', 'Test AUC'],
    Stacking_mnb_rf: ['Train Accuracy', 'Train Macro Precision', 'Train Macro Recall', 'Train Macro F1-Score', 'Train AUC', 'Test Accuracy', 'Test Macro Precision', 'Test Macro Recall', 'Test Macro F1-Score', 'Test AUC'],
    Stacking_rf_lr: ['Train Accuracy', 'Train Macro Precision', 'Train Macro Recall', 'Train Macro F1-Score', 'Train AUC', 'Test Accuracy', 'Test Macro Precision', 'Test Macro Recall', 'Test Macro F1-Score', 'Test AUC'],
    Support_vector_machine: ['Train Accuracy', 'Train Macro Precision', 'Train Macro Recall', 'Train Macro F1-Score', 'Train AUC', 'Test Accuracy', 'Test Macro Precision', 'Test Macro Recall', 'Test Macro F1-Score', 'Test AUC'],
    XGboost: ['Train Accuracy', 'Train Macro Precision', 'Train Macro Recall', 'Train Macro F1-Score', 'Train AUC', 'Test Accuracy', 'Test Macro Precision', 'Test Macro Recall', 'Test Macro F1-Score', 'Test AUC']
}
```

```python
# Initialize lists to store mean and standard deviation values for each metric (training and testing)
mean_train_accuracy = []
std_train_accuracy = []
mean_test_accuracy = []
std_test_accuracy = []

mean_train_precision = []
std_train_precision= []
mean_test_precision = []
std_test_precision = []

mean_train_recall = []
std_train_recall = []
mean_test_recall = []
std_test_recall = []

mean_train_f1score = []
std_train_f1score = []
mean_test_f1score = []
std_test_f1score = []

mean_train_auc = []
std_train_auc = []
mean_test_auc = []
std_test_auc = []
```

```python
# Iterate through each CSV file
for file in Files:
    # Load CSV into DataFrame
    df = pd.read_csv(file)
    
    # Get metrics columns for this file
    cols = metrics_columns[file]
    
    # Calculate mean and std
    means, stds = calculate_stats(df, cols)
    
    # Append mean and std accuracy to lists
    mean_train_accuracy.append(means['Train Accuracy'])
    std_train_accuracy.append(stds['Train Accuracy'])
    mean_test_accuracy.append(means['Test Accuracy'])
    std_test_accuracy.append(stds['Test Accuracy'])
    
    # Append mean and std precision to lists
    mean_train_precision.append(means['Train Macro Precision'])
    std_train_precision.append(stds['Train Macro Precision'])
    mean_test_precision.append(means['Test Macro Precision'])
    std_test_precision.append(stds['Test Macro Precision'])
    
    # Append mean and std recall to lists
    mean_train_recall.append(means['Train Macro Recall'])
    std_train_recall.append(stds['Train Macro Recall'])
    mean_test_recall.append(means['Test Macro Recall'])
    std_test_recall.append(stds['Test Macro Recall'])
    
    # Append mean and std F1-score to lists
    mean_train_f1score.append(means['Train Macro F1-Score'])
    std_train_f1score.append(stds['Train Macro F1-Score'])
    mean_test_f1score.append(means['Test Macro F1-Score'])
    std_test_f1score.append(stds['Test Macro F1-Score'])
    
    # Append mean and std AUC to lists
    mean_train_auc.append(means['Train AUC'])
    std_train_auc.append(stds['Train AUC'])
    mean_test_auc.append(means['Test AUC'])
    std_test_auc.append(stds['Test AUC'])
```

```python
# Plotting training and testing accuracy values
fig, ax = plt.subplots(figsize=(10, 8))

# Set positions for scatter points
x_pos = np.arange(len(Files))

# Plot scatter points with error bars
scatter = ax.errorbar(x_pos, mean_train_accuracy, yerr=std_train_accuracy, fmt='o', linestyle='-', color='red',markersize=8, capsize=5)
scatter = ax.errorbar(x_pos, mean_test_accuracy, yerr=std_test_accuracy,fmt='o', linestyle='-', color='darkblue', markersize=8, capsize=5)

# Add labels, title, and custom x-axis tick labels
ax.set_xlabel('Algorithms',fontsize=18)
ax.set_ylabel('Accuracy', fontsize=18)
ax.tick_params(axis='x', labelsize=16)
ax.tick_params(axis='y', labelsize=16)
ax.set_xticks(x_pos)
ax.set_ylim(0.985, 1.001)
ax.set_xticks(range(len(x_pos)), ['AdaBoost', 'DT', 'KNN', 'RF', 'Stacking(MNB_RF)', 'Stacking(RF_LR)', 'SVM', 'XGBoost'], rotation=45, fontsize=16, ha='center')
ax.grid(True)

# Show plot
plt.tight_layout()
plt.savefig('Accuracy.png', dpi=1000, bbox_inches='tight')
plt.show()
```
